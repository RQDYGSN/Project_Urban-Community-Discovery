### Node Embedding —— Graph Convolutional Network

```python
gcn_msg = fn.copy_src(src='h', out='m')
gcn_reduce = fn.sum(msg='m', out='h')
A = torch.from_numpy(adj).float()
d = torch.from_numpy(np.sum(adj, axis=1).reshape((-1, 1))).float()
mm = np.sum(adj)
B = A - torch.mm(d, d.T)/mm
B_norm = B/mm
DG1 = dgl.DGLGraph()
DG1.from_networkx(G)
features = torch.from_numpy(feature).float()

# D_ = th.diag(th.pow(d.view(-1), -1/2))
# A_norm = torch.mm(th.mm(D_, A), D_)

k = 15
in_dim = features.size()[1]
hid_dim = int(np.sqrt(in_dim*k))
```

```python
class GCNLayer(nn.Module):
    def __init__(self, in_feats, out_feats):
        super(GCNLayer, self).__init__()
        self.linear = nn.Linear(in_feats, out_feats)

    def forward(self, g, feature):
        # Creating a local scope so that all the stored ndata and edata
        # (such as the `'h'` ndata below) are automatically popped out
        # when the scope exits.
        with g.local_scope():
            g.ndata['h'] = feature
            g.update_all(gcn_msg, gcn_reduce)
            h = g.ndata['h']
            return self.linear(h)
        
class Net(nn.Module):
    def __init__(self, B):
        super(Net, self).__init__()
        self.layer1 = GCNLayer(in_dim, hid_dim)
        self.layer2 = GCNLayer(hid_dim, k)
        self.B = B

    def forward(self, g, features):
        global gcn_output_1
        global gcn_output_2
        global U
        n = nn.Softmax(dim=1)
        
        gcn_output_1 = self.layer1(g, features)
        # x = n(gcn_output_1)
        x = F.relu(gcn_output_1)
        # x = F.relu(self.layer1(g, features))
        gcn_output_2  = self.layer2(g, x)
        weight_feature = n(gcn_output_2)
        U = weight_feature.T
        H = torch.mm(torch.mm(weight_feature.T, self.B), weight_feature) # weight_feature 就是我们要求的U
        # Hh = torch.div(H, H.sum(axis=0)) # 按照每一列进行归一化，列为axis=0的方向。
        m = nn.Softmax(dim=0)
        # return m(Hh)
        return m(H)
net = Net(B_norm)
print(net)

def lossFn(output): 
    return torch.trace(-torch.log(output))
```

```python
optimizer = torch.optim.Adam(net.parameters(), lr=1e-2)
# optimizer = torch.optim.Adam(net.parameters(), lr=1e-2, weight_decay=0, betas=(0.5, 0.999), eps=1e-08)
epoch_num = 600
dur_loss = []
dur_modu = []
dur = []

net.train()

for epoch in range(epoch_num):
    if epoch >=0:
        t0 = time.time()
    
    # net.train()
    output = net(DG1, features)
    # Z_norm = th.div(Z, Z.sum(axis=1).view(-1, 1))
    # U = F.softmax(Z, dim=1).T
    # L = torch.mm(th.mm(U, B_norm), U.T)
    # loss = -torch.trace(L)
    loss = lossFn(output)
    dur_loss.append(loss)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if epoch >=0:
        dur.append(time.time() - t0)

    partition = dict(zip(list(G.nodes()), torch.argmax(U, axis=0).tolist())) 
    dur_modu.append(community.modularity(partition, G, weight='weight'))
    if epoch%20 == 0:
        print("Epoch {:05d} | Loss {:.4f} | Modularity {:.4f} | Time(s) {:.4f}".format(epoch, loss.item(), dur_modu[-1], np.mean(dur)))
```
